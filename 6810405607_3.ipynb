3.1 บริษัทให้บริการด้าน IT ครบวงจรแห่งหนึ่งได้เก็บรวบรวมข้อมูลค่าซ่อมบำรุงอุปกรณ์คอมพิวเตอร์แม่ข่ายและจำนวนลูกค้าของบริษัทในปีที่ผ่านมาเป็นดังไฟล์ asm3_1_itcost.csv  
จงแสดงวิธีหาค่ามัธยฐานและฐานนิยมของค่าซ่อมบำรุงอุปกรณ์คอมพิวเตอร์แม่ข่าย

```python
df_it = pd.read_csv('007.csv')
df_it.head()
```

```python
# หามัธยฐาน
df_it['total_num'] = df_it['num_customer'].cumsum()
display(df_it)
med = df_it['num_customer'].sum()/2
print(med)
mi = (5001 + 5000) / 2
print(mi)
summi = mi + ((med - df_it.total_num[0])/df_it.num_customer[1])*7
print(summi)

#หาฐานนิยม
h = df_it['num_customer'].idxmax()
de_m = np.abs(df_it.num_customer[h] - df_it.num_customer[h-1])
de_m1 = np.abs(df_it.num_customer[h] - df_it.num_customer[h+1])
summp = mi + (de_m/(de_m+de_m1))*7
print(summp)

```

3.2 สุ่มตัวอย่างนักศึกษามาจำนวนหนึ่ง พบว่าคะแนนสอบวิชาสถิติ เป็นดังตาราง 'asm3_2_score.csv' จงหาค่า

**3.2.1 ค่าเฉลี่ย**

```python
df_score = pd.read_csv('https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/asm3_2_score.csv')
df_score.head()
```

```python
df_score[['mi','mx']] = df_score['score'].str.split('-',expand=True).astype(int)
df_score['mid'] = (df_score.mx + df_score.mi) / 2
df_score['fx'] = df_score['num_student'] * df_score['mid']
b = df_score['fx'].sum() / df_score['num_student'].sum()
print(b)
display(df_score)
```

**3.2.2 ค่ามัธยฐาน**

```python
df_score['total_num'] = df_score['num_student'].cumsum()
display(df_score)
med = df_score['num_student'].sum() / 2
print(med)
mii = (25 + 26) / 2
print(mii)
summ = mii + ((med - df_score.total_num[2])/ df_score.num_student[3])*6
print(summ)
```

**3.2.3 ค่าฐานนิยม**

```python
tp = df_score['num_student'].idxmax()
de = df_score.num_student[tp] - df_score.num_student[tp-1]
de1 = df_score.num_student[tp] - df_score.num_student[tp+1]
mi = (df_score.mi[tp] - df_score.mx[tp-1]) / 2;
sumi = mi + (de / (de + de1))*6 
print(sumi)
```

**3.2.4 Q<sub>1</sub> และ Q<sub>3</sub>**

```python
#q1
n = 100
q1_1 = n*1/4
display(df_score)
print(q1_1)
e = 4
l = (df_score.mx[e-1] + df_score.mi[e]) / 2
q1 = l + ((q1_1 - df_score.total_num[e-1])/df_score.num_student[e])*5
print('q1 =',q1)


#q3
n = 100
q3_1 = n*3/4
print(q3_1)
e = 4
l = (df_score.mx[e-1] + df_score.mi[e]) / 2
q3 = l + ((q3_1 - df_score.total_num[e-1])/df_score.num_student[e])*5
print('q3 =',q3)
```

**3.2.5 D<sub>7</sub>**

```python
display(df_score)
n = 100
d1_1 = n*7/10
print(d1_1)
e = 4
l = (df_score.mx[e-1] + df_score.mi[e]) /2

d7 = l + ((d1_1 - df_score.total_num[e-1])/df_score.num_student[e])*5
print('d7 =',d7)
```

**3.2.6 P<sub>75</sub>**

```python
display(df_score)
n = 100
p1_1 = n*75/100
print(p1_1)
e = 4
l = (df_score.mi[e] + df_score.mx[e-1])/2

p75 = l + ((p1_1 - df_score.total_num[e-1])/df_score.num_student[e])*5
print('p75 = ',p75)
```

**3.2.7 พิสัย**

```python
display(df_score)
p =  df_score['mx'].iloc[-1] - df_score['mi'].iloc[0]
print('พิสัย =',p)
```

**3.2.8 ส่วนเบี่ยงเบนควอร์ไทล์**

```python
qd = (q3 - q1) / 2
print('Q.D =',qd)
```

**3.2.9 ส่วนเบี่ยงเบนมาตราฐานและความแปรปรวน**

```python
# เบี่ยงเบนมาตราฐาน 
display(df_score)
x = sum((df_score['mi']+df_score['mx'])/2 * df_score['num_student']/ df_score['num_student'].sum())
n = df_score['num_student'].sum()
print(n)
fx2 = sum(((df_score['mi']+df_score['mx']) /2) ** 2 * df_score['num_student'])
s = np.sqrt((fx2 - n*x**2) / (n-1))
print("เบี่ยงเบนมาตราฐาน =",s)
#ความแปรปรวน
s2 = s**2
print("แปรปรวน = ",s2)
```

**3.2.10 สัมประสิทธิ์ของการแปรผัน**

```python
cv = (s / x ) * 100
print(cv)
```

3.3 บริษัทเกมแห่งหนึ่งได้ทำการทดสอบ game engine รุ่นใหม่ ในคอมพิวเตอร์ที่มีส่วนประกอบของฮารด์แวร์ต่างๆกัน (เช่น cpu, gpu, ram, ฯลฯ) จำนวน 30 เครื่อง และทำการวัดค่า frames per second (FPS) ได้ผลดังนี้  
60.0, 30.2, 33.4, 59.9, 40.1, 25.5, 17.4, 28.6, 45.5, 57.8,  
22.5, 50.2, 30.8, 43.1, 50.5, 35.5, 27.4, 29.2, 55.5, 10.2,

จงคำนวณหาค่าเฉลี่ย, ส่วนเบี่ยงเบนมาตราฐาน, ฐานนิยม, Q<sub>1</sub>, D<sub>3</sub>, P<sub>45</sub> ของค่า FPS ที่ได้จากการทดสอบ

ค่าเฉลี่ย:

```python
fps = [60.0, 30.2, 33.4, 59.9, 40.1, 25.5, 17.4, 28.6, 45.5, 57.8,
       22.5, 50.2, 30.8, 43.1, 50.5, 35.5, 27.4, 29.2, 55.5, 10.2,
       32.2, 27.8, 33.8, 53.4, 25.5, 30.4, 36.1, 49.3, 18.6, 50.0]
```

```python
may = np.mean(fps)
print(may)
```

ส่วนเบี่ยงเบนมาตราฐาน:

```python
std = np.std(fps, ddof=1)
print(std)
```

ฐานนิยม:

```python
mo = statistics.mode(fps)
print(mo)
```

Q<sub>1</sub>, D<sub>3</sub> และ P<sub>45</sub>:  

```python
q1_2 = np.percentile(fps, 25)
print("q1 = ",q1_2)
d3_2 = np.percentile(fps, 30)
print("d3 = ",d3_2)
p45_2 = np.percentile(fps, 45)
print("p45 = ",p45_2)
```
