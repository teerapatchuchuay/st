```python
li1 = [10, 17, 25, 13, 21, 35, 15, 24, 17, 28]
me = np.mean(li1)
s = np.std(li1,ddof=1);
u = me + stats.t.ppf(0.1/2,df=10-1) * s/np.sqrt(10)
l = me + stats.t.isf(0.1/2,df=10-1) * s/np.sqrt(10)
print(u,l)
```
